# Import all library

```python
from eegrecord import cvxEDA
import numpy as np
import pylab as pl
```

# Define folder path

```python
path = '/content/eegrecord'
```

# Process the data

## 1. Data Latihan

### Import the data

```python
data = np.loadtxt("{}/Data_Latihan.txt".format(path), delimiter='\t',usecols = 0)
```

### Process the data

```python
y = data
yn = (y-y.mean()) / y.std()
Fs = 104
[r, p, t, l, d, e, obj] = cvxEDA.cvxEDA(yn, 1./Fs)
```

### Plot the data

```python
tm = pl.arange(1., len(y)+1.) / Fs
pl.plot(tm, yn)
```

```python
pl.plot(tm, r)
```

```python
pl.plot(tm, p)
```

```python
pl.plot(tm, t)
```

### Combine raw data and processed data

```python
allData = np.stack((tm, data, yn, r, p , t), axis = 1 )
```

### Save the data to csv

```python
np.savetxt("{}/cvxLatihan.csv".format(path), allData, delimiter=",")
```
